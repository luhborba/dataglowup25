```python
# Importando Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Explorando Dados

```python
# Carregando dados
df = pd.read_csv('lol_ranked_games.csv')
df.head()
```

```python
# Verificando Schema
df.info()
```

```python
# Verificando contagem de valores do gameId
contagem_gameId = df['gameId'].value_counts()
print(contagem_gameId)
```

```python
# Criando campo de KDA, Verificando se o nÃºmero de mortes Ã© igual a 0 para nÃ£o criar valores com erro
df['kda'] = np.where(df['deaths'] == 0, (df['kills'] + df['assists']), (df['kills'] + df['assists']) /( df['deaths'] ))
```

```python
df.head(10)
```

```python
# Reduzindo nÃºmero de casas decimais
df['kda']  = df['kda'].round(2)
df.head()
```

```python
# Criando novo DF com dados que serÃ£o apenas utilizados
df_final = df[[
    'gameId',
    'hasWon',
    'goldDiff',
    'champLevelDiff',
    'isFirstTower',
    'isFirstBlood',
    'killedElderDrake',
    'killedBaronNashor',
    'kills',
    'deaths',
    'assists',
    'wardsPlaced',
    'wardsDestroyed',
    'wardsLost',
    'kda'
]]
```

```python
df_final.head()
```

```python
df_final.info()
```

# AnÃ¡lises Comparativas

```python
# Filtrando o DataFrame para quando o time venceu (hasWon == 1)
vitorias = df_final[df_final['hasWon'] == 1]

# Filtrando o DataFrame para quando o time perdeu (hasWon == 0)
derrotas = df_final[df_final['hasWon'] == 0]
```

## Calculando Comparativo de KDA 

```python
# Calculando a mÃ©dia do KDA para vitÃ³rias e derrotas
media_kda_vitorias = vitorias['kda'].mean()
media_kda_derrotas = derrotas['kda'].mean()


# Exibindo os resultados
print("MÃ©dia do KDA para vitÃ³rias:", media_kda_vitorias.round(2))
print("MÃ©dia do KDA para derrotas:", media_kda_derrotas.round(2))
```

## Calculando Comparativo de Wards Colocadas

```python
# Calculando a mÃ©dia de wardsPlaced para vitÃ³rias e derrotas
media_wardsPlaced_vitorias = vitorias['wardsPlaced'].mean()
media_wardsPlaced_derrotas = derrotas['wardsPlaced'].mean()

# Exibindo os resultados
print("MÃ©dia de wardsPlaced para vitÃ³rias:", media_wardsPlaced_vitorias.round(2))
print("MÃ©dia de wardsPlaced para derrotas:", media_wardsPlaced_derrotas.round(2))
```

## Calculando Comparativo de Wards DestruÃ­das

```python
# Calculando a mÃ©dia de wardsDestroyed para vitÃ³rias e derrotas
media_wardsDestroyed_vitorias = vitorias['wardsDestroyed'].mean()
media_wardsDestroyed_derrotas = derrotas['wardsDestroyed'].mean()

# Exibindo os resultados
print("MÃ©dia de wardsDestroyed para vitÃ³rias:", media_wardsDestroyed_vitorias.round(2))
print("MÃ©dia de wardsDestroyed para derrotas:", media_wardsDestroyed_derrotas.round(2))
```

## Calculando Comparativo de Gold Diff (DiferenÃ§a de Gold)

```python
# Calculando a mÃ©dia de goldDiff para vitÃ³rias e derrotas
media_goldDiff_vitorias = vitorias['goldDiff'].mean()
media_goldDiff_derrotas = derrotas['goldDiff'].mean()

# Exibindo os resultados
print("MÃ©dia de goldDiff para vitÃ³rias:", media_goldDiff_vitorias.round(2))
print("MÃ©dia de goldDiff para derrotas:", media_goldDiff_derrotas.round(2))
```

# AnÃ¡lises Gerais

## Calculando Chance de VitÃ³ria atravÃ©s do campo KDA 

```python
# Calculando o valor mÃ©dio do KDA
valor_medio_kda = df_final['kda'].mean()

# Filtrando as partidas que o KDA estÃ¡ acima da mÃ©dia
partidas_kda_maior_media = df_final[df_final['kda'] > valor_medio_kda]
```

```python
# Contando o nÃºmero de vitÃ³rias e derrotas nas partidas em que o KDA Ã© acima da mÃ©dia
vitorias_kda_maior_media = partidas_kda_maior_media[partidas_kda_maior_media['hasWon'] == 1]
derrotas_kda_maior_media = partidas_kda_maior_media[partidas_kda_maior_media['hasWon'] == 0]

# Calculando a chance de vitÃ³ria
chance_vitoria_kda_maior_media = len(vitorias_kda_maior_media) / len(partidas_kda_maior_media)
```

```python
# Exibindo os resultados

print("Valor mÃ©dio do KDA:", valor_medio_kda.round(2))
print('-----------------')
print("NÃºmero de partidas com KDA acima da mÃ©dia:", len(partidas_kda_maior_media))
print('-----------------')
print("NÃºmero de vitÃ³rias com KDA acima da mÃ©dia:", len(vitorias_kda_maior_media))
print('-----------------')
print("NÃºmero de derrotas com KDA acima da mÃ©dia:", len(derrotas_kda_maior_media))
print('-----------------')
print("Chance de vitÃ³ria com KDA acima da mÃ©dia: {:.2f}".format(chance_vitoria_kda_maior_media))
```
